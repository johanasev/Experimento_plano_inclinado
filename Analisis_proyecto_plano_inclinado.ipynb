# **Aplicamos coeficiente de determinación R2**

El coeficiente de determinación (R2) proporciona una medida de cuánta variabilidad en la variable dependiente (en este caso, la aceleración) puede ser explicada por la variable independiente (el tiempo) a través del modelo lineal ajustado. Un valor de R
2 cercano a 1 indica que el modelo lineal explica una gran parte de la variabilidad en los datos, mientras que un valor cercano a 0 indica que el modelo no explica bien la variabilidad de los datos.

Para interpretar el R2 obtenido:

Si R2 es cercano a 1 (por ejemplo, **R2 ≥ 0.9**), podríamos concluir que el modelo lineal proporciona un buen ajuste a los datos experimentales y que la aceleración de la esfera en el plano inclinado puede ser explicada en gran medida por el tiempo.

Si R2 es moderado (por ejemplo, **0.7 ≤ R2 <0.9**), podríamos concluir que el modelo lineal es aceptable pero podría haber otros factores o variables que también afecten la aceleración y no están considerados en el modelo actual.

Si R2 es bajo (por ejemplo, **R2 < 0.7**), podríamos concluir que el modelo lineal no es un buen ajuste para los datos y que la aceleración de la esfera no puede ser bien explicada por el tiempo en el experimento, lo que sugiere la necesidad de considerar un modelo más complejo o tener en cuenta otros factores que puedan influir en la aceleración.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Datos proporcionados
data = {
    't': [0.03007, 0.05768, 0.08907, 0.115, 0.143, 0.172, 0.201, 0.232, 0.26,
          0.288, 0.316, 0.343, 0.372, 0.402, 0.428],
    'ax': [2.638, 2.986, 2.988, 2.766, 2.198, 1.769, 2.49, 2.319, 2.284,
           2.078, 2.324, 2.368, 2.14, 1.61, np.nan]
}

# Crear un DataFrame con los datos
df = pd.DataFrame(data)

# Eliminar filas con valores NaN
df = df.dropna()

# Ajuste de mínimos cuadrados
coefficients = np.polyfit(df['t'], df['ax'], 1)
polynomial = np.poly1d(coefficients)
df['ax_fit'] = polynomial(df['t'])

# Calcular el coeficiente de determinación (R²)
r_squared = r2_score(df['ax'], df['ax_fit'])

# Gráfico de dispersión y línea de mejor ajuste
plt.figure(figsize=(8, 6))
plt.scatter(df['t'], df['ax'], label='Datos experimentales', color='blue')
plt.plot(df['t'], df['ax_fit'], label=f'Línea de mejor ajuste: {coefficients[0]:.4f}t + {coefficients[1]:.4f}\n$R^2$ = {r_squared:.4f}', color='red')
plt.title('Gráfico de dispersión con mínimos cuadrados')
plt.xlabel('t')
plt.ylabel('ax')
plt.ylim(0, 6)  # Establecer límites del eje y
plt.legend()
plt.grid(True)
plt.show()

print(f"Coeficiente de determinación (R²): {r_squared:.4f}")
```
